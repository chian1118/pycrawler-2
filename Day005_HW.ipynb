```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

'''
Your Code
'''
import requests
import json
res = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data = json.loads(res.text)

print ("ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™:%s" % (len(data)))

for i, ele in enumerate(data):
    print ("ç¬¬%sç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½:" % (i+1), ele.keys())
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

'''
Your Code
'''
for i, ele in enumerate(data):
    print ("ç¬¬%sç­†è³‡æ–™" % (i+1))
    print ('æ¨™é¡Œ', ele['title'])
    print ('è²¼æ–‡æ™‚é–“', ele['createdAt'])
    print ('ç•™è¨€äººæ•¸', ele['commentCount'])
    print ('æŒ‰è®šäººæ•¸', ele['likeCount'])
    print ('---------------------------')
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

'''
Your Code
'''

commentCount = 0
likeCount = 0

for i, ele in enumerate(data):
    commentCount += ele['commentCount']
    likeCount += ele['likeCount']
    
print ('å¹³å‡ç•™è¨€äººæ•¸', commentCount / len(data))
print ('å¹³å‡æŒ‰è®šäººæ•¸', likeCount / len(data))
```
