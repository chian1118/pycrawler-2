## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

'''
Your Code
'''
import requests
import json
res = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data = json.loads(res.text)

print ("一次會回傳幾筆資料:%s" % (len(data)))

for i, ele in enumerate(data):
    print ("第%s筆資料包含哪些欄位:" % (i+1), ele.keys())
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

'''
Your Code
'''
for i, ele in enumerate(data):
    print ("第%s筆資料" % (i+1))
    print ('標題', ele['title'])
    print ('貼文時間', ele['createdAt'])
    print ('留言人數', ele['commentCount'])
    print ('按讚人數', ele['likeCount'])
    print ('---------------------------')
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

'''
Your Code
'''

commentCount = 0
likeCount = 0

for i, ele in enumerate(data):
    commentCount += ele['commentCount']
    likeCount += ele['likeCount']
    
print ('平均留言人數', commentCount / len(data))
print ('平均按讚人數', likeCount / len(data))
```
