### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
'''
Your Code
'''

# 讀檔案
import xmltodict

with open("./Data/64_72hr_CH.xml", "r") as fh:
    xml = fh.read()

print("File I/O:", type(xml))

doc = dict(xmltodict.parse(xml))
print (type(doc))
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

'''
Your Code
'''
count = 0
for i in doc['cwbopendata']['dataset']['locations']['location']:
    for j in i['weatherElement']:
        if j['description'] == '溫度':
            count += 1
            
print (count)
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

'''
Your Code
'''
for i in doc['cwbopendata']['dataset']['locations']['location']:
    for j in i['weatherElement']:
        tempature = []
        period = []
        if j['description'] == '溫度':
            times = j['time']
            period.append(times[0]['dataTime'])
            tempature.append(times[0]['elementValue']['value'])
            print ("%s所記錄的第一個時間點-%s跟溫度:%s" % (i['locationName'], period[0], tempature[0]))      
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

'''
Your Code
'''

for num, i in enumerate(doc['cwbopendata']['dataset']['locations']['location']):
    if num == 0:
        print (i['locationName'])
        for j in i['weatherElement']:
            if j['description'] == '溫度':
                times = j['time']
                for ele in times:
                    print (ele['dataTime'], ele['elementValue']['value'])
   
```
