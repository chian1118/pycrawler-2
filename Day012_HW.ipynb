### ① 取出今天所有的發文

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup
url = 'https://www.ettoday.net/news/news-list.htm'
res = requests.get(url)
soup = BeautifulSoup(res.text, "lxml")
```

```python
print (soup.find(class_="part_list_2"))
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
data = []
for ele in soup.find(class_="part_list_2").find_all('h3'):
    data.append({'Date':ele.find(class_="date").text,
                 'Category':ele.find('em').text,
                 'Title': ele.find_all('a')[-1].text,
                 'ContentUrl':'https://www.ettoday.net' + ele.find_all('a')[-1]['href']})
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
import pandas as pd

df = pd.DataFrame(data)
df.groupby(['Category'])['Title'].count().reset_index(name='count').sort_values(['count'], ascending=False)
```
