## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

## 1. 取出今天所有的新聞

```python
'''
Your Code
'''

from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime
import time

browser = webdriver.Chrome(executable_path='chromedriver')
browser.implicitly_wait(5)
browser.get("https://www.ettoday.net/news/news-list.htm")


for i in range(100):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "lxml")

data = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    data.append({'NewsDate': datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M'),
                 'NewsCategory':d.find('em').text,
                 'NewsContentUrl':'https://www.ettoday.net' + d.find_all('a')[-1]['href'],
                 'NewsTitle': d.find_all('a')[-1].text})
    
data
```

## 2. 取出現在時間兩小時內的新聞

```python
'''
Your Code
'''
import pandas as pd
from datetime import timedelta

max_time = datetime.now()
min_time = datetime.now() - timedelta(hours = 2) 

df = pd.DataFrame(data)
df[(df['NewsDate'] >= min_time) & (df['NewsDate'] <= max_time)]
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
'''
Your Code
'''
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
import time

browser = webdriver.Chrome(executable_path='chromedriver')
browser.implicitly_wait(5)
browser.get("https://www.ettoday.net/news/news-list.htm")


selectSite = Select(browser.find_element_by_id("selD"))
selectSite.select_by_value('15')

browser.find_element_by_id('button').click()

for i in range(70):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "lxml")


data = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    data.append({'NewsDate': datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M'),
                 'NewsCategory':d.find('em').text,
                 'NewsContentUrl':'https://www.ettoday.net' + d.find_all('a')[-1]['href'],
                 'NewsTitle': d.find_all('a')[-1].text})
    
min_time = datetime.strptime('2020/03/15 15:00', '%Y/%m/%d %H:%M')
max_time = datetime.strptime('2020/03/15 17:00', '%Y/%m/%d %H:%M')
    
df = pd.DataFrame(data)
df[(df['NewsDate'] >= min_time) & (df['NewsDate'] <= max_time)]
```
